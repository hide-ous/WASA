```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, HTML

from crawler.http_utils import http_ratelimited
```

```python
fpath = 'lesson_plan.html'
```

```python
with open(fpath, 'w', encoding='utf8') as f:
    f.write(http_ratelimited(requests.get, 'https://corsidilaurea.uniroma1.it/it/course/33503/attendance/lessons-plan').text)
```

```python
# 1. Parsing dell'HTML
with open(fpath, encoding='utf8') as f:
    soup = BeautifulSoup(f, 'html.parser')
```

```python
for p in soup.find_all('p'):
    print(p.prettify())
    break
```

```python
help(soup.find)
```

```python
class_name='curriculum'


# 2. Trova l'elemento con la classe specificata (curriculum)
curriculum_div = soup.find('div', class_=class_name)

if not curriculum_div:
    print(f"ERRORE BS4: Elemento con classe '{class_name}' non trovato.")

```

```python
curriculum_div = soup.find(lambda tag: tag.has_attr('class') and class_name in tag['class'])
```

```python
print(curriculum_div.prettify())
```

```python
display(HTML(curriculum_div.prettify()))
```

```python
curriculum_divs = soup.find_all(lambda tag: tag.has_attr('class') and class_name in tag['class'])
```
